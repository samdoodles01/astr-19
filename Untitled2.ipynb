```python
# Cell 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cell 2: Create a numpy array x ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Cell 3: Define functions, return sin(x) and cos(x)
def sine_function(x):
    return np.sin(x)

def cosine_function(x):
    return np.cos(x)
```

```python
# Cell 4: Use matplotlib to create a plot
y_sin = sine_function(x)
y_cos = cosine_function(x)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left panel: sin(x) vs x
axes[0].plot(x, y_sin, label="sin(x)", color='blue')
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
axes[0].set_title("sin(x) vs x")
axes[0].grid(True)
axes[0].legend()

# Right panel: cos(x) vs x
axes[1].plot(x, y_cos, label="cos(x)", color='red')
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")
axes[1].set_title("cos(x) vs x")
axes[1].grid(True)
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Save the figure as a PDF
plt.savefig("sin_cos_plot.pdf")

# Show the plot
plt.show()
```
