```python
# Cell 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cell 2: 1000 random numbers using Beta distribution
# set parameters a=2.0 and b=5.0
random_numbers = np.random.beta(a=2.0, b=5.0, size=1000)
```

```python
# Cell 3: histogram with 100 bins then save as PDF
plt.figure(figsize=(8, 5))
plt.hist(random_numbers, bins=100, color='mediumseagreen', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Beta-Distributed Random Numbers (a=2.0, b=5.0)")
plt.grid(True)
plt.tight_layout()
plt.savefig("beta_distribution_histogram.pdf")
plt.show()
```
